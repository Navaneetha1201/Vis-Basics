```python
from numpy import sin
from matplotlib import pyplot
x = [x*0.1 for x in range(50)]
y = sin(x)
pyplot.plot(x, y)
pyplot.show()
```

```python
from numpy.random import seed
from numpy.random import randn
from matplotlib import pyplot
seed(1)
x = 20 * randn(1000) + 100
y = x + (10 * randn(1000) + 50)
pyplot.scatter(x, y)
pyplot.show()
```

```python
from random import seed
from random import randint
from matplotlib import pyplot
seed(1)
x = ['red', 'green', 'blue']
y = [randint(10, 100), randint(0, 100), randint(0, 100)]
pyplot.bar(x, y)
pyplot.show()
```
